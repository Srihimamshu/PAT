```python
def climbStairs(n):
    def fib(n):
        if n==0 or n==1:
            return 1
        return fib(n-1)+fib(n-2)
    if n<=1:
        return 1
    for i in range(n):
        return fib(n)
print(climbStairs(40))
```

```python
def stoneGame(p):
    n = len(p)
    dp = [[0] * n for i in range(n)]
    for i in range(n): dp[i][i] = p[i]
    for d in range(1, n):
        for i in range(n - d):
            dp[i][i + d] = max(p[i] - dp[i + 1][i + d], p[i + d] - dp[i][i + d - 1])
    return dp[0][-1] > 0
p=[5, 3, 4, 5]
print(stoneGame(p))
```

```python
#DELETION IN BINARY TREE

# Algorithm

# 1.Start from the root of the tree.
# 2.If the tree is empty,return None
# 3.If the root node is the node to be deleted:
#         If the root has no children, return None
#         If the root has only one child, return the child.
#         If the root has two children:
#             Find the deepest node in the right subtree (or left subtree).
#             Copy the data of the deepest node to the root.
#             Delete the deepest node.
# 4.Otherwise, perform a level order traversal to find the node to be deleted.
# 5.Once the node to be deleted is found:
#         If the node has no childer,delete the node.
#         If the node has only one child, replace the node with its child
#         If the node has two childern:
#             Find the deepest node in the right subtree(or left subtree).
#             Copy the data of the deepest node to the root.
#             Delete the deepest node.
# 6.Return the root of the modified tree.

class Node:
    def __init__(self,data):
        self.data=data
        self.left=None
        self.right=None
def inorder(root):
    if not root:
        return
    inorder(root.left)
    print(root.data,end=" ")
    inorder(root.right)
def deleted_deepest(root,d_node):
    q=[root]
    while q:
        temp=q.pop(0)
        if temp is d_node:
            temp=None
            return 
        if temp.right:
            if temp.right is d_node:
                temp.right=None
                return 
            else:
                q.append(temp.right)
        if temp.left:
            if temp.left is d_node:
                temp.left=None
                return
            else:
                q.append(temp.left)
def deletion(root,key):
    if root is None:
        return None
    if root.left is None and root.right is None:
        if root.data==key:
            return None
        else:
            return root
        key_node=None
        q=[root]
        temp=None
        while q:
            temp=q.pop(0)
            if temp.data==key:
                key_node=temp
            if temp.left:
                q.append(temp.left)
            if temp.right:
                q.append(temp.right)
        if key_node:
            x=temp.data
            delete_deepest(root,temp)
            key_node=x
        return root

if __name__=="__main__":
    root=Node(50)
    root.left=Node(30)
    root.right=Node(70)
    root.left.left=Node(20)
    root.left.right=Node(40)
    root.right.left=Node(60)
    root.right.right=Node(80)
    
    print("Tree before deletion: ",end=" ")
    inorder(root)
    
    key=30
    root=deletion(root,key)
    
    print("Tree after deletion: ",end=" ")
    inorder(root)
```
